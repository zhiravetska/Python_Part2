```python
import pandas as pd

df = pd.read_csv("spotify_song_data.csv")

print(df)
```

```python
list(df.columns)
```

```python
df_sort = df.sort_values(by='track_popularity', ascending=False)
df_sort.head()
```

```python
shazamed_popularity = df_sort[['track_name', "artist_name", "album", "track_popularity",'total_shazamed_count']]
shazamed_popularity
```

```python
len(shazamed_popularity)
```

```python
import seaborn as sns
sns.jointplot(data=shazamed_popularity, x="track_popularity", y="total_shazamed_count", kind='reg')
```
