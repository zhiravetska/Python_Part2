```python
## Map&Filter
import random
```

```python
list_of_numbers=[]
n=100
for i in range(n):
    list_of_numbers.append(random.randint(2,99))
print(list_of_numbers)
```

```python
def double(x):
    return 2 * x
```

```python
result = map(double, list_of_numbers)
```

```python
list_of_doubled_numbers = list(result)
print(list_of_doubled_numbers)
```

```python
new_list_of_doubled_numbers = list(map(lambda element: element,list_of_doubled_numbers))## the second variant
```

```python
print(new_list_of_doubled_numbers)
```

```python
odd_numbers = [num for num in list_of_numbers if num % 2 != 0]## list comprehen.
print(odd_numbers)
```

```python
odd_numbers = list(filter(lambda num: num % 2 != 0, list_of_numbers))##filter function
print(odd_numbers)
```

```python
odd_numbers = (num for num in list_of_numbers if num % 2 != 0)##generator exp.
print(list(odd_numbers))
```

```python
##Defaultdict 
from collections import defaultdict
```

```python
def list_of_words(words):
    def_dict = defaultdict(list)
    for word in words:
        if word:
            first_ch = word[0]
            def_dict[first_ch].append(word)
            
    return def_dict
    
```

```python
words = ['tiger', 'dog', 'fish', 'elephant', 'cat', 'squrrel', 'donkey', 'penguin']
char_word_dict = list_of_words(words)
```

```python
char_word_dict['d']
```

```python
def get_country_info(country_input):
    
    name = ""
    try:
        name = country_input["name"]["common"]
    except:
        pass
 
    capital = ""
    try:
        capital = country_input["capital"]
    except:
        pass
   

    return name, capital
    

countries_data = list(map(get_country_info,countries))
```
