```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth 
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
from getpass import getpass
```

```python
#Authentication

cid = getpass("Client ID: ")
secret = getpass("Client Secret: ")
```

```python
client_credentials_manager = SpotifyClientCredentials(client_id=cid, client_secret=secret)
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
```

```python
playlist_link = "https://open.spotify.com/playlist/37i9dQZEVXbNG2KDcFcKOF"
playlist_URI = playlist_link.split("/")[-1].split("?")[0]
track_uris = [x["track"]["uri"] for x in sp.playlist_tracks(playlist_URI)["items"]]
```

```python
track_uris
```

```python
trackuri = []
tname = []
artisturi = []
artistinfo = []
artname = []
artpop = []
artgenres = []
albumname = []
tpop = []
```

```python
for track in sp.playlist_tracks(playlist_URI)["items"]:
    track_uri = track 
    artist_uri = track["track"]["artists"][0]["uri"]
    artist_info = sp.artist(artist_uri)
    artist_name = track["track"]["artists"][0]["name"]
    artist_pop = artist_info["popularity"]
    artist_info["genres"]
    album = track["track"]["album"]["name"]
    track_pop = track["track"]["popularity"]

artist_name 
artist_pop 
artist_info 
album 
track_pop 
```

```python
for track in sp.playlist_tracks(playlist_URI)["items"]:
    trackuri.append(track["track"]["uri"])
    tname.append(track["track"]["name"])
    artisturi.append(track["track"]["artists"][0]["uri"])
    artistinfo.append(sp.artist(artist_uri))
    artname.append(track["track"]["artists"][0]["name"])
    artpop.append(artist_info["popularity"])
    artgenres.append(artist_info["genres"])
    albumname.append(track["track"]["album"]["name"])
    #tpop.append(track["track"]["popularity"])
    
```

```python
data = {"Track_URI": trackuri,
        "Track Name": tname,
        "Artist URI": artisturi,
        "Artist Info": artistinfo,
        "Artist Name": artname, 
        "Artist Pop": artpop, 
        "Genres": artgenres,
        "Album": albumname}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
acoustic = []
dance = []
energy = []
instrumental = []
liveness = []
loudness = []
speech = []
tempo = []
valence = []
popularity = []
duration = []


for i in track_uris:
    feat = sp.audio_features(i)[0]
    acoustic.append(feat['acousticness'])
    dance.append(feat['danceability'])
    energy.append(feat['energy'])
    speech.append(feat['speechiness'])
    instrumental.append(feat['instrumentalness'])
    loudness.append(feat['loudness'])
    tempo.append(feat['tempo'])
    liveness.append(feat['liveness'])
    valence.append(feat['valence'])
    popu = sp.track(i)
    popularity.append(popu['popularity'])
    duration.append(feat['duration_ms'])

   
```

```python
X = ari_sort_pop[["dance",
                "acoustic",
                "energy",
                "instrumental",
                "liveness", 
                "speech",
                "tempo",
                "valence",
                "duration_ms"]]

y = ari_sort_pop["track_popular"]
import statsmodels.api as sm
model = sm.OLS(y,X)
results = model.fit()
results.pvalues
#results.summary()
```

Danceability, acoustic, speech, tempo and duration can be considered as significant. 

```python
results.summary()
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure(data=go.Scatter(x=X,y=y,mode="markers",name="pop"))
```

```python
fig.add_scatter(x=X,y=results.fittedvalues,name="Model")
```

```python
# R-squared is 0.973, indicating a perfect fit of the model, p-values of acoustic,
# energy, instrumental, liveness, speech, tempo, valence are > 0.05, these parameters are insignificant. 
# Then we'll repeat analysis excluding these factors.

X = ari_sort_pop[["dance",
                  "energy"]]

y = ari_sort_pop["track_popular"]
import statsmodels.api as sm
model = sm.OLS(y,X)
results = model.fit()
results.pvalues
#results.summary()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import avg

spark = SparkSession.builder.appName("AverageDuration").getOrCreate()

data = spark.read.options(delimiter=",").csv("ari_sort_pop.csv", header=True, inferSchema=True)
average_scores = data.groupBy("Artist Name").agg(avg("duration_ms").alias("AverageDuration"))

average_scores.show()
spark.stop()
```

```python
spark = SparkSession.builder.appName("Popularity").getOrCreate()

data = spark.read.options(delimiter=",").csv("ari_sort_pop.csv", header=True, inferSchema=True)
average_scores = data.groupBy("Artist Name").agg(avg("track_popular").alias("Popularity"))

average_scores.show()
spark.stop()
```
